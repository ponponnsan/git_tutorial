```
# 外れ値の除去
def outlier_2s(df):

    for i in range(len(df.columns)):

        # 列を抽出する
        col = df.iloc[:,i]

        # 平均と標準偏差
        average = np.mean(col)
        sd = np.std(col)

        # 外れ値の基準点
        outlier_min = average - (sd) * 2
        outlier_max = average + (sd) * 2

        # 範囲から外れている値を除く
        col[col < outlier_min] = None
        col[col > outlier_max] = None

    return df
```

```
# randomforest
def rf(df):
# モデルの定義
  # 不要な説明変数を削除
  x, y = df.drop(["cnt","atemp","yr","id"], axis=1), df["cnt"]
  # 訓練⽤データセットとテスト⽤データセットへの分割
  x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
  rfr = RandomForestRegressor(n_estimators=100, random_state=1)
  
  # モデルの訓練
  # 訓練データセットの説明変数と⽬的変数を引数に渡す
  rfr.fit(x_train,y_train)

  #精度
  print('学習用のデータ精度：{:.3f}'.format(rfr.score(x_train, y_train) ))
  print('テスト用のデータ精度：{:.3f}'.format(rfr.score(x_test, y_test) ))

  # X_train 特徴量
  importances=rfr.feature_importances_
  indices=np.argsort(importances)[::-1]
  for f in range((x_train.shape[1])):
    print("%2d) %-*s %f "%(f+1,30,x_train.columns[indices[f]],importances[indices[f]]))
```
